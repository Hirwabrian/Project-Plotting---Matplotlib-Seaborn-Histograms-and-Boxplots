```python
# YOUR CODE HERE
plt.figure(figsize=(10,6))
sns.histplot(df['co2'], bins=20, kde=True)
plt.title("Distribution of CO₂ Emissions")
plt.xlabel("CO₂ Emissions (g/mile)")
plt.ylabel("Frequency")
plt.grid(True)
plt.savefig("hist_co2.png")
plt.show()
```

```python
# YOUR CODE HERE(other 2 histograms)
plt.figure(figsize=(10,6))
plt.hist(df['displ'], bins=15, color='#e74c3c', edgecolor='black', alpha=0.8)
# Styling
plt.title("Engine Displacement Distribution", pad=15)
plt.xlabel("Engine Size (Liters)", labelpad=10)
plt.ylabel("Number of Vehicles", labelpad=10)
plt.grid(axis='y', alpha=0.4)
plt.savefig("hist_Eng_Disp.png")
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(10,6))
ax = sns.histplot(df['comb'], bins=25, kde=True,
                 color='#3498db', stat='percent')
plt.title("Fuel Efficiency Distribution", pad=15)
plt.xlabel("Combined MPG", labelpad=10)
plt.ylabel("Percentage of Vehicles", labelpad=10)
sns.despine()
plt.savefig("hist_Fuel_Effi.png")
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(12,8))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Features")
plt.savefig("heatmap_corr.png")
plt.show()
```

```python
#interpretation
#The heatmap shows fuel economy metrics (city/highway/comb MPG) are strongly correlated.
#They measure similar performance aspects.
#CO₂ emissions inversely correlate with efficiency scores (higher emissions = lower ratings).
#Vehicle ID and year show no meaningful relationships.
#Engine specs (cylinders/displacement) impact both emissions and efficiency but are redundant with each other.
# Identifiers like 'id' and weak predictors like 'year' should be excluded and would be the one's i'd delete as they are "non-predictive".
# Either cylinders or displ could be deleted since they are highly correlated (correlation ≈ 0.93) and this helps avoid redudancy.
```
